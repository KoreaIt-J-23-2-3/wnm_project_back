```python
from selenium import webdriver # 크롬 드라이버
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd
import time
from bs4 import BeautifulSoup
import json

# dict > list > dict
url = "https://marlonshop.com/"
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
driver = webdriver.Chrome(options=chrome_options)
driver.get(url)

# 페이지 소스 가져오기
# html = driver.page_source

# # soup에 넣어주기
# soup = BeautifulSoup(html, 'html.parser')
# print(soup.text)
time.sleep(4)
driver.find_element(by=By.XPATH, value='//*[@id="wrap"]/header/div/div[2]/div[1]/div/div/ul/li[1]/a').click();
time.sleep(4)

data = {}
data_list = []


# 고양이 카테고리 -> 0=홈리빙 2=이동 4=장난감 3=패션
cat_category_list = ["/web/upload/brushlab/img/ctgIcon/cat_home.png", "/web/upload/brushlab/img/ctgIcon/cat_carrier.png", "/web/upload/brushlab/img/ctgIcon/cat_toy.png", "/web/upload/brushlab/img/ctgIcon/cat_acc.png"]

# 강아지 카테고리 -> 0=홈리빙 1=산책 2=이동 3=패션 4=장난감  value=f'a[herf="dog_category_list[{0}]"]'
dog_category_list = ["/web/upload/brushlab/img/ctgIcon/dog_home.png", "/web/upload/brushlab/img/ctgIcon/dog_walk.png", "/web/upload/brushlab/img/ctgIcon/dog_carrier.png", "/web/upload/brushlab/img/ctgIcon/dog_clothes.png", "/web/upload/brushlab/img/ctgIcon/dog_toy.png"]


driver.find_element(by=By.CSS_SELECTOR, value=f'a[href="/product/list.html?cate_no=255"]').click();
time.sleep(4)

for j in range(1,6):
    driver.find_element(by=By.CSS_SELECTOR, value=f'img[src="{dog_category_list[j - 1]}"]').click();
    time.sleep(3)

    driver_list = []
    driver_selected_elements = driver.find_elements(by=By.CSS_SELECTOR, value='li[class="xans-record-"]');

    for driver_selected_element in driver_selected_elements:
        driver_list.append(driver_selected_element.get_attribute('id'))
    time.sleep(1)

    for i in range(1, 21):
        product_url = driver.find_element(by=By.CSS_SELECTOR, value=f'li[id="{driver_list[i - 1]}"] div[class="thumbnail"] a')
        product_href = product_url.get_attribute("href");

        driver.get(product_href)
        element = driver.page_source;

        soap = BeautifulSoup(element, 'html.parser')

        content_div = soap.find('div', class_='simple_desc')
        inner_text = ''.join([str(item) for item in content_div.contents if item.name is None]) 


        time.sleep(4)
        price = driver.find_element(by=By.CSS_SELECTOR, value=f'div[id="totalProducts"] table tfoot tr td span strong em').text
        price_data = price.replace(",", "").replace("원", "") 

        thumb_img = driver.find_element(by=By.CSS_SELECTOR, value=f'img[class="ThumbImage"]').get_attribute("src")
        detail_img = driver.find_element(by=By.CSS_SELECTOR, value=f'div[class="jq-tab-content active"] img').get_attribute("src")

        data = {
            "productName": driver.find_element(by=By.CSS_SELECTOR, value='div[class="infoArea"] div h3').text,
            "productPrice": price_data,
            "productDetailText": inner_text,
            "productThumbnail": thumb_img,
            "productDetailImg": detail_img,
            "petTypeId": 0,
            "categoryId": j - 1,
            "noSize": 0,
            "productSizeXS": 0,
            "productSizeS": 0,
            "productSizeM": 0,
            "productSizeL": 0,
            "productSizeXL": 0,
            "productSizeXXL": 0
        }
        data_list.append(data)
        driver.back();
        time.sleep(3)
    driver.back();
    time.sleep(2)






driver.find_element(by=By.CSS_SELECTOR, value=f'a[href="/product/list.html?cate_no=256"]').click();
time.sleep(4)


for j in range(1,5):
    # 0=홈리빙 2=이동 4=장난감 3=패션

    k = 0
    if(j == 1):
        k = 2
    elif(j == 2):
        k = 4
    elif(j == 3):
        k = 3

    driver.find_element(by=By.CSS_SELECTOR, value=f'img[src="{cat_category_list[j - 1]}"]').click();
    time.sleep(3)

    driver_list = []
    driver_selected_elements = driver.find_elements(by=By.CSS_SELECTOR, value='li[class="xans-record-"]');

    for driver_selected_element in driver_selected_elements:
        driver_list.append(driver_selected_element.get_attribute('id'))
    time.sleep(1)


    for i in range(1, 21):
        product_url = driver.find_element(by=By.CSS_SELECTOR, value=f'li[id="{driver_list[i - 1]}"] div[class="thumbnail"] a')
        product_href = product_url.get_attribute("href");

        driver.get(product_href)
        element = driver.page_source;

        soap = BeautifulSoup(element, 'html.parser')

        content_div = soap.find('div', class_='simple_desc')
        inner_text = ''.join([str(item) for item in content_div.contents if item.name is None]) 


        time.sleep(4)
        price = driver.find_element(by=By.CSS_SELECTOR, value=f'div[id="totalProducts"] table tfoot tr td span strong em').text
        price_data = price.replace(",", "").replace("원", "") 

        thumb_img = driver.find_element(by=By.CSS_SELECTOR, value=f'img[class="ThumbImage"]').get_attribute("src")
        detail_img = driver.find_element(by=By.CSS_SELECTOR, value=f'div[class="jq-tab-content active"] img').get_attribute("src")

        data = {
            "productName": driver.find_element(by=By.CSS_SELECTOR, value='div[class="infoArea"] div h3').text,
            "productPrice": price_data,
            "productDetailText": inner_text,
            "productThumbnail": thumb_img,
            "productDetailImg": detail_img,
            "petTypeId": 1,
            "categoryId": k,
            "noSize": 0,
            "productSizeXS": 0,
            "productSizeS": 0,
            "productSizeM": 0,
            "productSizeL": 0,
            "productSizeXL": 0,
            "productSizeXXL": 0
        }
        data_list.append(data)
        driver.back();
        time.sleep(3)
    driver.back();
    time.sleep(2)

with open("product_json_data.json", "w", encoding="utf-8") as f:
    json.dump(data_list, f, indent=4, ensure_ascii=False)
```
